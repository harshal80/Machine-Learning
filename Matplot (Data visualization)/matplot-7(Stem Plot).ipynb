# Stem Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y)
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='r',basefmt='g',label='python')

plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='r',basefmt='g',label='python'  ,use_line_collection=False)

plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt='r',basefmt='g',label='python'  ,use_line_collection=False,orientation="horizontal")

plt.legend()
plt.show()
```
