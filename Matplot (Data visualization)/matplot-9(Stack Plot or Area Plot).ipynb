# Stack Plot or Area Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area=[2,3,3,5,4]

plt.stackplot(x,area)
```

```python
x=[1,2,3,4,5]
area1=[2,3,3,5,4]
area2=[1,3,6,2,5]
area3=[6,3,1,6,4]

l=['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l)

plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,3,5,4]
area2=[1,3,6,2,5]
area3=[4,3,1,6,4]

l=['area1','area2','area3']


plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','g','m'],baseline='wiggle')

plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,3,5,4]
area2=[1,3,6,2,5]
area3=[6,3,1,6,4]

l=['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,labels=l)

plt.xlabel("python")
plt.ylabel("y-axis.")
plt.title("Harshal Bhau na visualization")


plt.legend(loc=2)
plt.show()
```
