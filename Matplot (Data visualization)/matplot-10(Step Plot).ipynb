# Step Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,1,4,5,3]

plt.grid()

plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Python")


plt.step(x,y)
```

```python
x=[1,2,3,4,5]
y=[2,1,4,5,3]

plt.grid()

plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Python")


plt.step(x,y,color='r',marker='o',ms=10,mfc='g',label='python')

plt.legend(loc=2)
plt.show()
```
