# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[3,2,4,6,7,5,4]

plt.xlabel("days",fontsize=13)
plt.ylabel("No.",fontsize=13)
plt.title("Scatter Plot",fontsize=13)

plt.scatter(x,y,color='r')
```

```python
x=[1,2,3,4,5,6,7]
y=[3,2,4,6,7,5,4]

colors=['b','g','b','r','y','r','g']
size=[100,20,30,50,60,10,40]

plt.xlabel("days",fontsize=13)
plt.ylabel("No.",fontsize=13)
plt.title("Scatter Plot",fontsize=13)

plt.scatter(x,y,c=colors,s=size)
```

```python
x=[1,2,3,4,5,6,7]
y=[3,2,4,6,7,5,4]

colors=['b','g','b','r','y','r','g']
size=[100,20,30,50,60,10,40]

plt.xlabel("days",fontsize=13)
plt.ylabel("No.",fontsize=13)
plt.title("Scatter Plot",fontsize=13)

plt.scatter(x,y,c=colors,s=size,alpha=0.6)
```

```python
x=[1,2,3,4,5,6,7]
y=[3,2,4,6,7,5,4]

colors=['b','g','b','r','y','r','g']
size=[100,20,30,50,60,10,40]

plt.xlabel("days",fontsize=13)
plt.ylabel("No.",fontsize=13)
plt.title("Scatter Plot",fontsize=13)

plt.scatter(x,y,c=colors,s=size,marker="*",edgecolor='r',linewidth=0.6)
```

```python
x=[1,2,3,4,5,6,7]
y=[3,2,4,6,7,5,4]

colors=['b','g','b','r','y','r','g']
size=[100,200,300,50,300,170,400]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,1,6,2]

plt.xlabel("days",fontsize=13)
plt.ylabel("No.",fontsize=13)
plt.title("Scatter Plot",fontsize=13)

plt.scatter(x,y,c=colors,s=size,cmap="viridis",alpha=0.5)
plt.scatter(x,no2,c='r',s=size,alpha=0.5)

t=plt.colorbar()
t.set_label('colorBar')
```
