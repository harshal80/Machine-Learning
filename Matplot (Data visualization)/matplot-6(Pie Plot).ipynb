# Pie Plot

```python
import matplotlib.pyplot as plt
```

```python
y=["Python","C","CPP","Java"]
x=[40,10,20,30]

plt.xlabel("Lanuage")

plt.title("Harshal Bhau na visualization")

plt.pie(x,labels=y)

plt.show()
```

```python
y=["Python","C","CPP","Java"]
x=[40,10,20,30]

plt.xlabel("Lanuage")

plt.title("Harshal Bhau na visualization")
ex=[0.0,0.3,0.0,0.0]

plt.pie(x,labels=y,explode=ex)

plt.show()
```

```python
y=["Python","C","CPP","Java"]
x=[40,10,20,30]

plt.xlabel("Lanuage")

plt.title("Harshal Bhau na visualization")
ex=[0.0,0.3,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow="True")

plt.show()
```

```python
y=["Python","C","CPP","Java"]
x=[40,10,20,30]

plt.xlabel("Lanuage")
plt.title("Harshal Bhau na visualization")
ex=[0.0,0.3,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow="True" ,radius=0.9,labeldistance=0.8)

plt.show()
```

```python
y=["Python","C","CPP","Java"]
x=[40,10,20,30]

plt.xlabel("Lanuage")
plt.title("Harshal Bhau na visualization")
ex=[0.0,0.3,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow="True" ,radius=0.9,labeldistance=0.8,startangle=180)

plt.show()
```

```python
y=["Python","C","CPP","Java"]
x=[40,10,20,30]

plt.xlabel("Lanuage")
plt.title("Harshal Bhau na visualization")
ex=[0.0,0.3,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow="True" ,radius=0.9,labeldistance=0.8 ,counterclock=False
        ,center=(3,6))

plt.show()
```

```python
y=["Python","C","CPP","Java"]
x=[40,10,20,30]

plt.xlabel("Lanuage")
plt.title("Harshal Bhau na visualization")
ex=[0.0,0.3,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow="True" ,radius=0.9,labeldistance=0.8,startangle=0
        ,counterclock=False,rotatelabels=False)

plt.legend(loc=3)
plt.show()
```
