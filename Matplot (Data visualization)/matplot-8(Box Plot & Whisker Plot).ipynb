# Box plot

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x)
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,widths=0.5)
```

```python
x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'],patch_artist=False,showmeans=True)
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,sym='g',boxprops=dict(color='r'),capprops=dict(color='g'),flierprops=dict(markeredgecolor='y'))
```

```python
x=[10,20,30,40,50,60,70,120]
x1=[10,20,40,90,30,50,80,60]
y=[x,x1]

plt.boxplot(y,labels=['python','C++'],showmeans=True,boxprops=dict(color='r'),capprops=dict(color='g')
            ,flierprops=dict(markeredgecolor='y'))
plt.show()
```
