# Savefig

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,1,5,3,4]

plt.plot(x,y,color='r')

plt.savefig("linePlot")

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,1,5,3,4]

plt.plot(x,y,color='r')

plt.savefig("linePlot1",dpi=2000)
#dpi:- provide the Quality of graph
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,1,5,3,4]

plt.plot(x,y,color='r')

plt.savefig("linePlot",facecolor='g')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,1,5,3,4]

plt.plot(x,y,color='r')

plt.savefig("linePlot.pdf")

#savefig(filename.pdf,png,jpg,....)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,1,5,3,4]

plt.plot(x,y,color='r')

plt.savefig("linePlot",facecolor='g',transparent=False,bbox_inches='tight')

plt.show()
```
