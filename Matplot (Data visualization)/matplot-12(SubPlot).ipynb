# SubPlot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,y,color='r')

x1=[10,20,30,40]

plt.subplot(2,2,2)
plt.pie([1],colors='g')

plt.subplot(2,2,3)
plt.pie(x1)

x2=['a','b','h','f']
y2=[10,20,30,40]


plt.subplot(2,2,4)
plt.bar(x2,y2)


plt.show()
```
