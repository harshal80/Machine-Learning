# Fill_Between

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.xlabel("X-axis")
plt.ylabel("Y-axis.")
plt.title('Python')

plt.plot(x,y,color='r')

plt.fill_between(x,y,color='g')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.xlabel("X-axis")
plt.ylabel("Y-axis.")
plt.title('Python')

plt.plot(x,y,color='r')

plt.fill_between(x=[2,4],y1=2,y2=4)

plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])



plt.xlabel("X-axis")
plt.ylabel("Y-axis.")
plt.title('Python')

plt.plot(x,y,color='r')

plt.fill_between(x,y,color='g',where=(x>=2)& (x<=4),alpha=0.5)

plt.show()
```
