# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]

plt.bar(x,y)
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]

plt.xlabel("Lanuage")
plt.ylabel("No.")
plt.title("Harshal Bhau na visualization")
plt.bar(x,y)
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]

plt.xlabel("Lanuage",fontsize=13)
plt.ylabel("No.",fontsize=13)
plt.title("Harshal Bhau na visualization",fontsize=13)
plt.bar(x,y)
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]

plt.xlabel("Lanuage")
plt.ylabel("No.")
plt.title("Harshal Bhau na visualization")

plt.bar(x,y,width=0.4)
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]

plt.xlabel("Lanuage")
plt.ylabel("No.")
plt.title("Harshal Bhau na visualization")

c=['y','b','m','g']

plt.bar(x,y,width=0.4,color=c)
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]

plt.xlabel("Lanuage")
plt.ylabel("No.")
plt.title("Harshal Bhau na visualization")

c=['y','b','m','g']

plt.bar(x,y,width=0.4,color=c,align="edge",edgecolor='r',linewidth=2,linestyle=':')
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]

plt.xlabel("Lanuage")
plt.ylabel("No.")
plt.title("Harshal Bhau na visualization")

c=['y','b','m','g']

plt.bar(x,y,width=0.4,color=c,align="edge",edgecolor='r',linewidth=2,linestyle=':',alpha=0.5)
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]

plt.xlabel("Lanuage")
plt.ylabel("No.")
plt.title("Harshal Bhau na visualization")

c=['y','b','m','g']

plt.bar(x,y,width=0.4,color=c,align="edge",edgecolor='r',label='Harsh')
plt.legend()
```

```python
x=["Python","C","CPP","Java"]
y=[88,30,80,90]
z=[30,20,53,25]

plt.xlabel("Lanuage")
plt.ylabel("No.")
plt.title("Harshal Bhau na visualization")

c=['y','b','m','g']

plt.bar(x,y,width=0.4,color="r",align="edge",label='Popularity1')
plt.bar(x,z,width=0.4,color="y",align="edge",label='Popularity2')

plt.legend()
```

```python
import numpy as np

x=["Python","C","CPP","Java"]
y=[88,30,80,90]
z=[30,20,53,25]

width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Lanuage")
plt.ylabel("No.")
plt.title("Harshal Bhau na visualization")


plt.bar(p,y,width,color="r",label='Popularity1')
plt.bar(p1,z,width,color="y",label='Popularity2')

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
